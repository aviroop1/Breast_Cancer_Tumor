<b>Smoothness_mean vs Area_mean values are quite different in scale. Do we normalize it?</b>

```python
#Normalize data
Diagnosis = BreastCancerData.iloc[:,1]
Features = BreastCancerData.iloc[:, 2:12]
Normalized = (Features - Features.mean())/Features.std()
NewData = pd.concat([Diagnosis, Normalized], axis = 1)
NewData = pd.melt(NewData, id_vars = "diagnosis", var_name = "features", value_name = "value")
```

```python
sns.violinplot(x = "features", y = "value", split = True, hue = "diagnosis", inner = "quart", data =  NewData)
plt.xticks(rotation = 90)
```

<b> Almost all the mean features can be accurately classified between malignant and benign, but features like fractal dimension and symmetry can't help us significantly distinguish betweeb malignant and benign tumors. This indicates that:
1. The malignant tumors are typically bigger in size (radius, perimeter, area) as compared to benign tumors
2. The malignant tumors measure higher in texture as compared to benign tumors. 
3. The malignant tumors are also smoother than benign tumors.
4. The malignant tumors are more compact than benign tumors.
5. The malignant and benign tumors have similar fractal dimensions and are symmetric.</b>
